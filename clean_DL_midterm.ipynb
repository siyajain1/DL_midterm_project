```python
import unsloth
from unsloth import FastLanguageModel
import torch

print("Torch:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("Device:", torch.cuda.get_device_name(0))


model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Meta-Llama-3.1-8B",
    max_seq_length=2048,
    dtype=None,
    load_in_4bit=True,
)
```

```python
from transformers import TrainingArguments
from trl import SFTTrainer
target_modules = [
    "q_proj", "k_proj", "v_proj", "o_proj",   # attention projection layers
    "gate_proj", "up_proj", "down_proj"       # MLP (feed-forward) layers
]
# --- LoRA configuration ---
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,                              # low-rank dimension (balance quality vs VRAM)
    lora_alpha = 32,                    # scaling factor
    target_modules = target_modules,  # focus on attention projections
    lora_dropout = 0.00,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
)


```

```python
from datasets import load_dataset

# Dataset ID
DATASET_ID = "ad6398/nyu-dl-teach-maths-comp"

# Load the full dataset
full_dataset = load_dataset(DATASET_ID, split="train")


train_subset = 500_000
valid_subset = 5_000

split_dataset = full_dataset.train_test_split(
    test_size = valid_subset / (train_subset + valid_subset),
    seed = 42,
)

train_dataset = split_dataset["train"].select(range(train_subset))
validation_dataset = split_dataset["test"].select(range(valid_subset))

print(f"Loaded {len(train_dataset):,} train / {len(validation_dataset):,} validation samples.")

# Define the prompt template — identical to the professor’s inference version
EOS_TOKEN = tokenizer.eos_token  # End of sequence marker

PROMPT_TEMPLATE = (
    "You are an expert mathematician evaluating whether a student's solution to a math question is correct. "
    "Read the question and the proposed solution carefully. Think step-by-step and respond only 'True' or 'False'.\n\n"
    "Question:\n{}\n\n"
    "Student's Solution:\n{}\n\n"
    "Your Evaluation (True/False):\n{}"
)

# Format function — includes is_correct label in the text
def formatting_prompts_func(batch):
    texts = []
    for q, s, o in zip(batch["question"], batch["solution"], batch["is_correct"]):
        prompt = PROMPT_TEMPLATE.format(q.strip(), str(s).strip(), str(o)) + EOS_TOKEN
        texts.append(prompt)
    return {"text": texts}

formatted_train_dataset = train_dataset.map(formatting_prompts_func, batched=True, num_proc=16)
formatted_validation_dataset = validation_dataset.map(formatting_prompts_func, batched=True, num_proc=16)


print(f"   {len(formatted_train_dataset)} training samples")
print(f"   {len(formatted_validation_dataset)} validation samples")
print("\nExample")
print(formatted_train_dataset[0]["text"][:2048])
```

```python
from transformers import TrainingArguments
from unsloth.trainer import SFTTrainer
output_dir = "/content/drive/MyDrive/llama3_math_competition_final"

training_args = TrainingArguments(
    output_dir = output_dir,
    per_device_train_batch_size = 8,
    gradient_accumulation_steps = 4,  # effective batch = 64
    num_train_epochs = 2,
    warmup_steps = 50,
    learning_rate = 5e-4,
    lr_scheduler_type = "cosine",
    logging_steps = 50,
    save_strategy = "steps",
    save_steps = 75,
    eval_strategy  = "steps",
    eval_steps = 75,
    report_to = "none",
    weight_decay = 0.05,
    bf16 = True,
    optim = "adamw_8bit",
    seed = 42,
)

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = formatted_train_dataset,
    eval_dataset = formatted_validation_dataset,
    dataset_text_field = "text",
    max_seq_length = 2048,
    packing = True,
    args = training_args,
)

print("Trainer initialized")

trainer.train()


model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
print(f"saved to: {output_dir}")
```

```python
import pandas as pd
from tqdm import tqdm
from datasets import load_dataset
import torch

# Load your fine-tuned model and tokenizer
from unsloth import FastLanguageModel
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "/content/drive/MyDrive/llama3_math_competition_final",
    max_seq_length = 2048,
    dtype = None,
    load_in_4bit = True,
)

# Load the competition test dataset
test_dataset = load_dataset("ad6398/nyu-dl-teach-maths-comp", split="test")

# Define prompt for inference
inference_prompt = """You are a great mathematician and you are tasked with finding if a solution to a given maths question is correct or not.
Your response should be 'True' if the solution is correct, otherwise 'False'.
Below is the Question and Solution.
Question:
{}
Solution:
{}
Output:
"""

# Helper: parse the model output into True/False
def parse_output(response_text):
    output_part = response_text.split("Output:")[-1]
    return 'true' in output_part.lower()

# Generate predictions
predictions = []
for example in tqdm(test_dataset):
    q, s = example["question"], example["solution"]
    prompt = inference_prompt.format(q, s)
    inputs = tokenizer([prompt], return_tensors="pt").to("cuda")

    outputs = model.generate(**inputs, max_new_tokens=8, use_cache=True)
    response = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
    predictions.append(parse_output(response))

# Save submission
submission = pd.DataFrame({
    "ID": range(len(predictions)),
    "is_correct": predictions
})
submission.to_csv("submission.csv", index=False)
print("submission.csv saved! Ready to upload to Kaggle.")
```
